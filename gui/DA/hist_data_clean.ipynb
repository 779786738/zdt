```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./cz_sh_hist_data_0823.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
col_list=['robot_id','sum_running_time','avg_inst_inst_temperature_1','sum_running_circles_1',
          'avg_current_fluctuation_1']
data_ans = pd.DataFrame(columns=col_list)

for i in range(1,7):
    col_list_temp = [col.replace('1',str(i)) for col in col_list]
    data_temp = data[col_list_temp]
    data_temp['joint'] = i
    data_temp.columns = col_list+['joint']
    data_ans = pd.concat([data_ans,data_temp])
```

```python
data_ans.iloc[2::44,:]
```

```python
data_ans = data_ans.rename(columns={'avg_inst_inst_temperature_1':'avg_inst_temperature','sum_running_circles_1':'sum_running_circles','avg_current_fluctuation_1':'avg_current_fluctuation'})
```

```python
data_ans['joint'] = data_ans['joint'].apply(np.int8)
```

```python
data_ans.to_csv('hist_data_0823.csv')
```
