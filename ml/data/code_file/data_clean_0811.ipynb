拆关节

```python
col_list=['robot_id','rtc','protective_stop','task_state','rapid_rate','sum_running_time',
          'ins_average_power_1','sum_running_circles_1','joint_1_actual_position',
          'current_fluctuation_1','inst_temperature_1','inst_voltage_1']
data_ans = pd.DataFrame(columns=col_list)
for i in range(1,7):
    col_list_temp = [col.replace('1',str(i)) for col in col_list]
    
    data_temp = data_clean[col_list_temp]
    data_temp['joint'] = i
    data_temp.columns = col_list+['joint']
    data_ans = pd.concat([data_ans,data_temp])
```
